CODIGO MODIFICADO - REGRESION LINEAL

```python
from ipywidgets import interact
import matplotlib.pyplot as plt

# Puntos iniciales
p1 = (5.4, 3.2)
p2 = (9.5, 0.7)
p3 = (12.3, -3.6)

def update_plot(p2_x, p2_y):
    p2 = (p2_x, p2_y)

    # Lista de las coordenadas X e Y
    x_coords = [p1[0], p2[0], p3[0]]
    y_coords = [p1[1], p2[1], p3[1]]

    # Clculo de las sumas necesarias para la regresin lineal
    n = len(x_coords)
    sum_x = sum(x_coords)
    sum_y = sum(y_coords)
    sum_x2 = sum(x**2 for x in x_coords)
    sum_xy = sum(x * y for x, y in zip(x_coords, y_coords))

    # Clculo de la pendiente (m) y la interseccin (b)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
    b = (sum_y - m * sum_x) / n


    plt.figure(figsize=(10, 6))
    plt.scatter(x_coords, y_coords, color="red")

    # Crear la lnea de regresin
    x_line = [min(x_coords), max(x_coords)]
    y_line = [m * x + b for x in x_line]
    plt.plot(x_line, y_line, color="blue")

    # Mostrar los detalles en la grfica
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Points and Line Plot")
    plt.show()

_ = interact(update_plot, p2_x=(5.5, 12.3, 0.1), p2_y=(-10.0, 10.0, 0.1))
```
